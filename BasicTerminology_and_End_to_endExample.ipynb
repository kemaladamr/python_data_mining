## Loading data into memory - viewing and managing with ease using pandas

```python
# load data into pandas
import pandas as pd
df = pd.read_csv('iris.csv')
```

```python
# sanity check with pandas
print('Shape of data in (rows, columns) is ' + str(df.shape))
print(df.head())
print(df.describe().transpose())
```

## Plotting and exploring data - harnessing the power of Seaborn

```python
# explore with Seaborn pairplot
import seaborn as sns
sns.pairplot(df, hue='species')
```

```python
# add histogram to diagonals of Seaborn pairplot
sns.pairplot(df, hue='species', diag_kind='hist',
             palette='bright',
             markers=['o', 'X', 'v']
             )
```

```python
# plot bivariate scatter with Seaborn
sns.lmplot(x='petal length in cm', y='petal width in cm',
           hue='species', data=df, fit_reg=False,
           palette='bright', markers=['o', 'x', 'v'])
```
